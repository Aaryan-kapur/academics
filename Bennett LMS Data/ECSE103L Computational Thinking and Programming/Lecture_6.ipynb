### Let one grain of wheat be placed on the first square of a chessboard, two grains on the second square, four grains on the third square, eight grains on the fourth square, and so on untill all square are filled in chessboard. what wiil be the total weight in ton of grains on whole 8×8 chessboard? If 15432 grains in one kg and 907.18 kg in one ton then.



#### series=  1+2+4+8+.......2pow64

```python
grains_in_one_ton = 15432*907.18
Total_grain=0
n=0
while(n<64):
    Total_grain+=2**n
    print(Total_grain, 'in', n,'th box')
    n=n+1
no_ton= Total_grain/grains_in_one_ton
print(no_ton,"Ton")
# 9700000 Ton Wheat produced in India per year
print(no_ton/97000000, 'years')
```

### write a program to find wether a given number is prime or not

```python
x = int(input("Enter a number"))
flag = True
for i in range(2,x):
    if(x%i==0):
        flag = False
if(flag):
    print("Number is Prime")
else:
    print("Number is not prime")
        
```
