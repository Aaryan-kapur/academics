### Find wether a given number is prime or not

```python
import time

x = 225
start = time.time()
flag = True
count=0
for i in range(2,x):
    count +=1
    if(x%i==0):
        flag = False
if(flag):
    print("Number is Prime")
else:
    print("Number is not prime")
end = time.time()
print("Number of iteration loop run- ",count)
print ("Total execution time", end-start)
```

### Better prime number program by checking till only half of given number

```python
import time
x = int(input("Enter a number"))
start = time.time()
flag = True
count=0
for i in range(2,(x//2 +1)):
    count +=1
    if(x%i==0):
        flag = False
if(flag):
    print("Number is Prime")
else:
    print("Number is not prime")
end = time.time()
print("Number of iteration loop run- ",count)
print ("Total execution time", end-start)
```

### Better prime number program by checking till only square root of given number

```python
import math
import time
x = int(input("Enter a number"))
start = time.time()
flag = True
count=0
y = int(math.sqrt(x))+1
for i in range(2,y):
    count +=1
    if(x%i==0):
        flag = False
if(flag):
    print("Number is Prime")
else:
    print("Number is not prime")
end = time.time()
print("Number of iteration loop run- ",count)
print ("Total execution time", end-start)
```
