### 5. Write a program for division of 2n and multiply 2n Using bitwise operator 

```python
number = int(input("Enter values of n: "))
division = number >> 1
multiplication = number << 1

print("Dividing a number by two is equal to the right shift by one bit :",division)
print("Multiplying a number by two is equal to the left shift by one :",multiplication)
```
